```python
import os
import cv2
import numpy as np
from pycocotools import coco, mask as maskUtils


# Define paths and filenames
data_dir = '/kaggle/working/valid'
ann_file = '/kaggle/working/valid/_annotations.coco.json'
img_dir = os.path.join(data_dir, 'images')

# Initialize COCO instance and load annotations
coco_inst = coco.COCO(ann_file)
ann_ids = coco_inst.getAnnIds()
anns = coco_inst.loadAnns(ann_ids)



j=0
for ann in list(anns[::]):
    # Extract segmentation coordinates
    if 'segmentation' in ann:
        # print(ann)
        segmentations = ann['segmentation']

        for i, seg in enumerate(segmentations):
            # Convert polygon to binary mask
            mask = coco_inst.annToMask(ann)

            # Create individual mask directory if it doesn't exist
            mask_dir = os.path.join(data_dir,str(ann["image_id"]),'masks')
            os.makedirs(mask_dir, exist_ok=True)

            # Save individual mask
            mask_filename = f'{ann["image_id"]}_{ann["category_id"]}_{j}.jpg'
            j=j+1
            mask_path = os.path.join(mask_dir, mask_filename)
            cv2.imwrite(mask_path, mask.astype(np.uint8) * 255)
```

```python
## Imports
import os
import sys
import random


import numpy as np
import cv2
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

## Seeding
seed = 42
random.seed = seed
np.random.seed = seed
tf.seed = seed
```

```python
import os
import cv2
import numpy as np
import keras.utils

class DataGen(keras.utils.Sequence):
    def __init__(self, ids, path, batch_size=16, image_size=320):
        self.ids = list(ids)  # Convert ids to a list for indexing and length calculation
        self.path = path
        self.batch_size = batch_size
        self.image_size = image_size
        self.on_epoch_end()  # Shuffle or reorganize data at the end of each epoch

    def __load__(self, id_name):
        try:
            # Construct paths for image and masks
            image_path = os.path.join(self.path, id_name, "images", id_name) + ".jpg"
            mask_path = os.path.join(self.path, id_name, "masks/")

            # Load and resize image
            image = cv2.imread(image_path, 1)
            if image is None:
                raise FileNotFoundError(f"Failed to load image: {image_path}")

            image = cv2.resize(image, (self.image_size, self.image_size))
            image = image.astype(np.float32) / 255.0  # Normalize to [0, 1]

            # Initialize mask
            mask = np.zeros((self.image_size, self.image_size, 1), dtype=np.float32)

            # Load and merge masks
            all_masks = os.listdir(mask_path)
            for name in all_masks:
                _mask_path = os.path.join(mask_path, name)
                _mask_image = cv2.imread(_mask_path, -1)
                _mask_image = cv2.resize(_mask_image, (self.image_size, self.image_size))
                _mask_image = np.expand_dims(_mask_image, axis=-1)
                mask = np.maximum(mask, _mask_image.astype(np.float32) / 255.0)  # Normalize to [0, 1]

            return image, mask
        except FileNotFoundError as e:
            print(e)
            return None, None

    def __getitem__(self, index):
        start = index * self.batch_size
        end = min((index + 1) * self.batch_size, len(self.ids))
        batch_ids = self.ids[start:end]

        images = []
        masks = []
        for id_name in batch_ids:
            image, mask = self.__load__(id_name)
            if image is not None and mask is not None:
                images.append(image)
                masks.append(mask)

        return np.array(images), np.array(masks)

    def on_epoch_end(self):
        np.random.shuffle(self.ids)  # Shuffle ids at the end of each epoch

    def __len__(self):
        return (len(self.ids) + self.batch_size - 1) // self.batch_size  # Equivalent to ceil(len(self.ids) / self.batch_size)
```

```python
import os

# Parameters
image_size = 320
train_path = "/kaggle/working/train"
epochs = 50
batch_size = 16

# Get all training IDs
train_ids = next(os.walk(train_path))[1]

# Limit to the first 100 images
train_ids = train_ids

# Validation Data Size
val_data_size = 52

# Ensure there are enough images for the validation set
if len(train_ids) < val_data_size:
    raise ValueError("Not enough images to create the validation set")

# Split into training and validation sets
valid_ids = train_ids[:val_data_size]
train_ids = train_ids[val_data_size:]

print(f"Training images: {len(train_ids)}")
print(f"Validation images: {len(valid_ids)}")
```

```python
gen = DataGen(train_ids, train_path, batch_size=batch_size, image_size=image_size)
x, y = gen.__getitem__(0)
print(x.shape, y.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Attention, Reshape
from tensorflow.keras.models import Model

def build_model(input_shape, num_classes):
    # Encoder
    input_layer = Input(shape=input_shape)
    conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(input_layer)
    pool1 = MaxPooling2D((2, 2), padding='same')(conv1)

    conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(pool1)
    pool2 = MaxPooling2D((2, 2), padding='same')(conv2)

    conv3 = Conv2D(128, (3, 3), activation='relu', padding='same')(pool2)
    pool3 = MaxPooling2D((2, 2), padding='same')(conv3)

    # Attention mechanism
    attention_input = Reshape((-1, 128))(pool3)
    attention_output = Attention()([attention_input, attention_input])
    attention_output = Reshape((80, 80, 128))(attention_output)

    # Bottleneck
    bottleneck = Conv2D(128, (3, 3), activation='relu', padding='same')(attention_output)

    # Decoder
    upsample1 = UpSampling2D((2, 2))(bottleneck)
    conv4 = Conv2D(128, (3, 3), activation='relu', padding='same')(upsample1)

    # Attention mechanism
    attention_output2 = Reshape((-1, 128))(conv4)
    attention_output2 = Attention()([attention_output2, attention_output2])
    attention_output2 = Reshape((160, 160, 128))(attention_output2)

    upsample2 = UpSampling2D((2, 2))(attention_output2)
    conv5 = Conv2D(64, (3, 3), activation='relu', padding='same')(upsample2)

    upsample3 = UpSampling2D((2, 2))(conv5)
    conv6 = Conv2D(32, (3, 3), activation='relu', padding='same')(upsample3)

    upsample4 = UpSampling2D((2, 2))(conv6)
    conv7 = Conv2D(16, (3, 3), activation='relu', padding='same')(upsample4)

    output_layer = Conv2D(num_classes, (3, 3), activation='sigmoid', padding='same')(conv7)

    model = Model(inputs=input_layer, outputs=output_layer)

    return model

# Define input shape and number of classes
input_shape = (640, 640, 3)  # Increased input size
num_classes = 1  # Binary segmentation

# Build and compile the model
model = build_model(input_shape, num_classes)
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["acc"])
model.summary()
```

```python
# Create data generators for training and validation
train_gen = DataGen(train_ids, train_path, batch_size=batch_size, image_size=image_size)
valid_gen = DataGen(valid_ids, train_path, batch_size=batch_size, image_size=image_size)  # Use the same path for validation

# Train the model
model.fit(train_gen, validation_data=valid_gen, epochs=epochs)
```
